```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/Job_Placement_Data.csv')
```

```python
df.head()
```

## 1. Distribution Plot

```python
sns.distplot(df['ssc_percentage'], kde=True, bins=5)
```

Interpretation: Highest number of student have a mark between 60-70. 

## 2. Bar Plot

```python
yt = pd.read_csv('./topSubscribed.csv')
yt.head()
```

```python
plt.figure(figsize=(8,6))
plt.xticks(rotation=90)
sns.barplot(x='Category',y='Subscribers',data=yt)
```

Interpret:
1. Shows and Trailers channel have highest average subscribers.
2. Rest Category have more or less same average subscribers. where travel&events have lowest average subscribers

## 3. Count Plot

```python
plt.xticks(rotation=90)
sns.countplot(x='Started',data=yt)
```

Interpret:
1. Most successful channel are created between 2011 to 2016.
2. As youtube founded on 2005. This is not possible to create channel before. So this also shows there are some error in this dataset.
3. Highest Succesfull channel started on 2014.

## 4. Heat Map

```python
yt_corr = yt.corr()
plt.figure(figsize=(5,3))
sns.heatmap(yt_corr,annot=True)
```

Interpret: 
1. There is strong correlation between Subscribers and Video views. and it is Positive. 
2. Though we may think more videos means more views, the table shows otherwise. There is a weak correlation between video count and views.

## 5. Pie Chart

```python
data = yt.groupby('Category').mean()

plt.pie(data['Video Views'], labels=data.index,autopct='%.0f%%')
plt.title('Views Percentage')
plt.show()
```

1. People spend around 16% time of total views on 'Shows' channel which highest by far. Closest competitors is 'Trailers' with 9% and 'Education' with 8%.
2. People spend less time on 'Howto & Style', 'Science & Technology' and 'Travel & Events' channel.
