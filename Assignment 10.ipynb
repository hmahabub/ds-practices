- create new column (column name 'initials') from Name column using only initials (mr./mrs./cap. etc are examples of initials and they end with a full stop ) 
- then fill missing Age value base on 'initials' column (using groupby mean)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/train.csv')
df.sample(3)
```

```python
df.shape
```

```python
# removing duplicate value if any
df.drop_duplicates()
```

```python
# calculating percentage of missing value
(df.isnull().sum() / df.shape[0]) * 100
```

```python
# getting initials from name column
def initial(row):        
    row['initials'] = [a for a in row.Name.split(' ') if a.endswith('.')][0]
    return row
new_df = df.apply(initial,  axis='columns')
new_df.sample(5)
```

```python
# reading groupwise mean age value
new_df.groupby('initials')['Age'].mean()
```

```python
new_df['Age'] = new_df.groupby('initials')['Age'].transform(lambda x: x.fillna(x.mean()))
```
